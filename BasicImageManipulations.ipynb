```python
import cv2 as cv
import matplotlib.pyplot as plt

# Carga la imagen, lo hace  mediante una matris 
imagen = cv.imread("MS_Pathfinder.jpg")

#tambien se puede cargar dando como parametro una variable con el phat

#path = "C:\Users\josesin\Desktop\MI CARPETA\Portafolio\Python\Programas\BasicImageManipulations\MS_Pathfinder.jpg"
#image = cv.imread(path)
```

```python
#cv2 lee la imagen en bgr y hay que trans formarla a rgb
imagenRGB = cv.cvtColor(imagen, cv.COLOR_RGB2BGR)
```

```python
# Mostrar la imagen abriendo una ventana nueva
#cv.imshow("Nombre Ventana",imagen)
#como se esta usando jupyter, la ventana queda a medias y es mejor usar matplot
plt.imshow(imagenRGB)

# Esperar indefinodo hasta que se presione una tecla cualquiera del teclado
#cv.waitKey(0)
```

```python
#tambien se pueden conseguir las dimensiones de la imagen de la forma (filas(alto),columnas(ancho),canles):
print (imagenRGB.shape)
```

```python
#se puedne cambiar las dimensiones de la imagen pero para ello hay que realizar algunos calculos para
#que este cambio sea proporcional y no distorcione la imagen 

#disminuir el ancho a la mitad 
relacion = 600 / imagenRGB.shape[1] #relacion entre el nuevo ancho y el original
dimension = ( 600, int ( imagenRGB.shape [ 0 ] * relacion )) 
#reconstruye las nuevas dimensiones donde el valore de la relacion es multiplicado por la altura original para encontrar
#la nueva altura, en el fondo es una reglade 3 :v
```

```python
#se realiza el cambio y se muestra la nueva imagen
imgResized = cv.resize(imagenRGB,dimension, interpolation = cv.INTER_AREA)

plt.imshow(imgResized)
```

```python
#comparar dimensiones
print (imagenRGB.shape)
print (imgResized.shape)
```

```python
#asi como se puede escalar la imagen, tambien se puede rotar
#para ello se obtienen las dimensiones y se almacenan en variables

(H,w) = imagenRGB.shape[:2]
#print(H,w)

#se calcula el centro, sera el punto donde se rotara la imagen
centro = (w/2,H/2)
#print (centro)
```

```python
#se declara una matriz para poder rotar la imagen
matrizRotacion = cv.getRotationMatrix2D(centro, 180, 1.0)
#cv2.getRotationMatrix2D (ptoDeRotacion,anguloDeRotacion,escala)
#print(matriz)
```

```python
#para realizar la rotacion se crea una nueva variable que almacene la imagen rotada
imgRotada = cv.warpAffine(imagenRGB,matrizRotacion,(w,H))
#cv.warpAffine(imagenParaRotar,Matris,Dimensiones)
```

```python
#motrar la imagen rotada
plt.imshow(imgRotada)
```

```python
#para recortar se le entrega los puntos en de eje en elplano dfe la imagen q se quieren mostras [startX : endX, startY : endY]
imgRecorte = imagenRGB[0:400,600:1200] 
#los valores los puse random xd, despues fui ajustando segun lo que queria ver
#aunque tome de referencia los pixeles :v 
plt.imshow(imgRecorte)
```

```python
#se puede guardar la imagen trabajada:
cv.imwrite("recortePathfinder.jpg",imgRecorte)
#cv.imwrite("nombreImagen.formato",imagenModificada)
```
